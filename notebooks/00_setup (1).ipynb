## Step 1 — Mount Drive & paths

```
from google.colab import drive
drive.mount('/content/drive')
import os
DRIVE_ROOT = '/content/drive/MyDrive/PICALIB_Research'
REPO_URL   = 'https://github.com/anasbiswas1/picalib-research.git'
REPO_DIR   = os.path.join(DRIVE_ROOT, 'picalib-research')
os.makedirs(DRIVE_ROOT, exist_ok=True)
print('Drive root:', DRIVE_ROOT)
```

## Step 2 — Git identity

```
!git config --global user.name  "Md Anas Biswas"
!git config --global user.email "anasbiswas@gmail.com"
print("identity set")
```

## Step 3 — Store token ONCE (persisted to Drive)

```
import getpass, os
TOKEN = getpass.getpass('Paste GitHub token (ghp_...): ')
cred = os.path.join(DRIVE_ROOT, '.git-credentials')
with open(cred, 'w') as f:
    f.write(f'https://anasbiswas1:{TOKEN}@github.com\n')
os.chmod(cred, 0o600)
!git config --global credential.helper "store --file={cred}"
del TOKEN; print('Token stored to Drive.')
```

## Step 4 — Clone (first time) or pull

```
if not os.path.exists(os.path.join(REPO_DIR, '.git')):
    %cd {DRIVE_ROOT}
    !git clone {REPO_URL}
else:
    %cd {REPO_DIR}
    !git pull
%cd {REPO_DIR}
!ls -la
```
