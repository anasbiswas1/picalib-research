## Step 4 — Store the token ONCE (persisted to Drive)
You paste your GitHub Personal Access Token (`ghp_...`) once. It is saved to Drive so future sessions never ask again. `getpass` keeps it out of the notebook output.

```
import getpass, os
TOKEN = getpass.getpass('Paste GitHub token (ghp_...): ')
cred = os.path.join(DRIVE_ROOT, '.git-credentials')
with open(cred, 'w') as f:
    f.write(f'https://anasbiswas1:{TOKEN}@github.com\n')
os.chmod(cred, 0o600)
!git config --global credential.helper "store --file={cred}"
del TOKEN
print('Token stored to Drive. You will not be asked again.')
```

## Step 5 — Clone (first time) or pull (later)

```
if not os.path.exists(os.path.join(REPO_DIR, '.git')):
    %cd {DRIVE_ROOT}
    !git clone {REPO_URL}
    print('Repo cloned.')
else:
    %cd {REPO_DIR}
    !git pull
    print('Repo already present — pulled latest.')
%cd {REPO_DIR}
!ls -la
```
